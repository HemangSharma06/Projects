# **CNN Model**

```python
import pandas as pd
import numpy as np
import tensorflow as tf
import json
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Flatten, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
window_size = 200
step = 100
segments = []
labels = []

feature_columns = df.columns.tolist()
feature_columns.remove('Type')
label_col = 'Type'

for tremor_type in sorted(df[label_col].unique()):
    subset = df[df[label_col] == tremor_type].reset_index(drop=True)
    for i in range(0, len(subset) - window_size, step):
        window = subset.iloc[i:i + window_size]
        segment_data = window[feature_columns].values
        segments.append(segment_data)
        labels.append(tremor_type)
```

```python
X = np.array(segments)
Y = np.array(labels)
```

```python
X_reshaped = X.reshape(-1, X.shape[-1])
scaler = StandardScaler()
X_scaled_reshaped = scaler.fit_transform(X_reshaped)
X_scaled = X_scaled_reshaped.reshape(X.shape)
```

```python
scaling_params = {
    'mean': scaler.mean_.tolist(),
    'scale': scaler.scale_.tolist()
}
scaling_params
```

```python
with open('scaling_values.json', 'w') as f:
    json.dump(scaling_params, f)
```

```python
encoder = OneHotEncoder(sparse_output=False)
y_encoded = encoder.fit_transform(Y.reshape(-1, 1))
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X_scaled, y_encoded, test_size=0.2, random_state=42, stratify=Y)
```

```python
model = Sequential([
    Conv1D(filters=64, kernel_size=3, activation='relu', input_shape=(X_train.shape[1], X_train.shape[2])),
    Conv1D(filters=64, kernel_size=3, activation='relu'),
    MaxPooling1D(pool_size=2),
    Dropout(0.5),

    Conv1D(filters=128, kernel_size=3, activation='relu'),
    Conv1D(filters=128, kernel_size=3, activation='relu'),
    MaxPooling1D(pool_size=2),
    Dropout(0.5),

    Flatten(),
    Dense(100, activation='relu'),
    Dense(Y_train.shape[1], activation='softmax')
])
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
```

```python
history = model.fit(
    X_train, Y_train,
    epochs=100,
    batch_size=32,
    validation_split=0.2,
    callbacks=[early_stopping],
    verbose=1
)
```

```python
converter = tf.lite.TFLiteConverter.from_keras_model(model)
converter.optimizations = [tf.lite.Optimize.DEFAULT]
tflite_model = converter.convert()
```

```python
tflite_filename = r"/content/drive/MyDrive/Final Project/earthquake_cnn_model.tflite"
with open(tflite_filename, 'wb') as f:
    f.write(tflite_model)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
plt.plot(history.history['accuracy'], label='accuracy')
plt.plot(history.history['val_accuracy'], label = 'val_accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.ylim([0, 1])
plt.legend(loc='lower right')
plt.show()

plt.plot(history.history['loss'], label='loss')
plt.plot(history.history['val_loss'], label = 'val_loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(loc='upper right')
plt.show()

y_pred_cnn = np.argmax(model.predict(X_test), axis=1)
Y_test_labels = np.argmax(Y_test, axis=1)
cm_cnn = confusion_matrix(Y_test_labels, y_pred_cnn)
cm_display_cnn = ConfusionMatrixDisplay(confusion_matrix=cm_cnn)
cm_display_cnn.plot(cmap="Reds")
plt.show()
```

```python
loss, accuracy = model.evaluate(X_train, Y_train, verbose=0)
print(f"CNN Model Accuracy on Training Data: {accuracy}")
```

```python
loss, accuracy = model.evaluate(X_test, Y_test, verbose=0)
print(f"CNN Model Accuracy on Test Data: {accuracy}")
```

```python
final_val_accuracy = history.history['val_accuracy'][-1]
print(f"CNN Model Final Validation Accuracy: {final_val_accuracy}")
```

```python
from sklearn.metrics import classification_report
import numpy as np

y_pred_cnn = np.argmax(model.predict(X_test), axis=1)
Y_test_labels = np.argmax(Y_test, axis=1)
print(classification_report(Y_test_labels, y_pred_cnn))
```
