```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split as TTS
from sklearn.ensemble import RandomForestClassifier as RFC
from sklearn.metrics import classification_report as cr
data = pd.read_csv("creditcard.csv")
x = data.iloc[:,:-1]
y = data['Class']
sm = SMOTE(random_state=42)
new_x, new_y = sm.fit_resample(x,y)
x_train,x_test,y_train, y_test = TTS(new_x, new_y, test_size=0.2, random_state=42) 
rf = RFC(n_estimators=25, n_jobs=-1,random_state=42)
rf.fit(x_train,y_train)
rf.score(x_test, y_test)*100
y_pred = rf.predict(x_test)
print(cr(y_test, y_pred))
from sklearn.metrics import ConfusionMatrixDisplay as CMD, confusion_matrix as cfm
mat = cfm(y_test, y_pred)
disp = CMD(confusion_matrix=mat, display_labels=rf.classes_)
disp.plot(cmap = plt.cm.Blues)
plt.show()
```

```python
# Step 1: Load and split before SMOTE
x_ = data.drop('Class', axis=1)
y_ = data['Class']
x_train_, x_test_, y_train_, y_test_ = TTS(x_, y_, test_size=0.2, random_state=42)

# Step 2: Apply SMOTE only to training data
smt = SMOTE(random_state=42)
x_train_res, y_train_res = smt.fit_resample(x_train_, y_train_)

# Step 3: Train and evaluate
rfc = RFC(n_estimators=100, n_jobs=-1, random_state=42)
rfc.fit(x_train_res, y_train_res)
print(f"Train Accuracy: {rfc.score(x_train_res, y_train_res):.4f}")
print(f"Test Accuracy: {rfc.score(x_test, y_test):.4f}")
```

```python
from sklearn.metrics import roc_auc_score, roc_curve

y_scores = rfc.predict_proba(x_test_)[:, 1]
auc = roc_auc_score(y_test_, y_scores)
print(f"Updated ROC AUC Score: {auc:.4f}")

fpr, tpr, _ = roc_curve(y_test_, y_scores)
plt.plot(fpr, tpr, label=f'AUC = {auc:.2f}')
plt.plot([0,1], [0,1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve (After Fixing SMOTE)')
plt.legend()
plt.show()
```
